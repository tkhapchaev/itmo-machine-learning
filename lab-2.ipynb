###  Описание данных
#### Таблица ```transactions.csv```
##### Описание
Таблица содержит историю транзакций клиентов банка за один год и три месяца.


````python
sample_size = 1000

sample = transactions["tr_type"].sample(n = sample_size, random_state = 1)
types_filtered = tr_types[tr_types["tr_description"].str.contains("АТМ") | tr_types["tr_description"].str.contains("POS")]
types_id = types_filtered["tr_type"].tolist()

sample.isin(types_id).value_counts().to_dict()[True] / sample_size
````

## Задание 2


1. Для столбца tr_type датафрейма transactions посчитайте частоту встречаемости всех типов транзакций tr_type в transactions.
2. Выведите топ-10 транзакций по частоте встречаемости (вывести для них tr_description тоже).

````python
frequency = transactions["tr_type"].value_counts(ascending = False)
frequency
````

````python
top_10 = list(frequency.head(10).to_dict())
result = pd.DataFrame(data = {"tr_type": [], "tr_description": [], "frequency": []})

for i in range(9):
  transaction_type = top_10[i]
  transaction_description = tr_types[tr_types["tr_type"] == transaction_type]["tr_description"].iloc[0]
  transaction_frequency = frequency[transaction_type]

  row = [transaction_type, str(transaction_description), transaction_frequency]

  result.loc[len(result)] = row
````

## Задание 3
1. В датафрейме transactions найдите клиента с максимальной суммой приходов на карту

3. Найдите модуль разницы для этих клиентов между суммой расходов и суммой приходов

````python
wastes = transactions[transactions["amount"] < 0]
wastes
````

````python
refills = transactions[transactions["amount"] >= 0]
refills
````

````python
top_refills = refills.groupby(["customer_id"])["amount"].sum().nlargest(n = 1)
top_refills
````

````python
top_wastes = wastes.groupby(["customer_id"])["amount"].sum().nsmallest(n = 1)
top_wastes
````

````python
difference = abs(abs(list(top_refills.to_list())[0]) - abs(list(top_wastes.to_list())[0]))
difference
````
